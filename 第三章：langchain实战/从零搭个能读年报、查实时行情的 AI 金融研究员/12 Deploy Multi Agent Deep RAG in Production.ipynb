### Core Tools
- **uv** - Python package manager (latest)
- **nvm** v0.40.3 (Linux/Mac) / **nvm-windows** (Windows)
- **Node.js** 24 LTS
- **Yarn** 4+
- **LangGraph CLI** 0.4.11

### System Requirements
- Ubuntu 20.04+ / Windows 10/11
- 8GB RAM, 10GB disk space
- Python 3.11+

## Installation Order

1. Update system
2. Install **uv**
3. Install **LangGraph CLI**
4. Install **nvm**
5. Install **Node.js 24**
6. Install **Yarn**
7. Configure API keys (.env)
8. Clone repositories
9. Install dependencies

## API Keys

- **LangSmith**: https://smith.langchain.com/ (required)
- **Google AI**: https://makersuite.google.com/app/apikey (optional)

## Repositories

- Backend: `github.com/laxmimerit/deep-finance-research`
- Frontend: `github.com/langchain-ai/deep-agents-ui`

## Quick Start

**Backend:**
```bash
cd deep-finance-research
langgraph dev  # http://localhost:2024
```

**Frontend:**
```bash
cd deep-agents-ui
yarn dev  # http://localhost:3000
```
